# Simple Linear Graph (Bank Robber vs Sheriff -- distance and velocity)

##### Given: 
- Robber travelling at 2.5 km/min
- Sheriff travelling at 3 km/min
- Sheriff starts 5 min after the robber

##### To find: 
- Point at which the sheriff catches the robber -> Intersection of the time=distance/velocity lines.

### Library imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Variable/object initialization

A 'linspace' object is created as a uniformly divided set of points to represent the time vector. 

```python
t = np.linspace(0, 40, 1000, endpoint=True)
```

Variable representing distance covered by robber

```python
d_r = 2.5 * t
```

Variable representing distance covered by sheriff

```python
d_s = 3 * (t - 5)
```

### Graphical solution
- Plotting superimposed graphs for distances covered by the sheriff and the robber
- The point of intersection represents the point where the robber is intercepted by the sheriff

### Method 1

##### Calling matplotlib.pyplot (as plt) to modify each attribute (not the preferred method)

```python
figure, axes = plt.subplots() # Creates a blank plot with figure and axes as return values that can be modified later

# Creating graph and axes labels
plt.title('Sheriff vs Bank Robber --- Distance vs time graph')
plt.xlabel('Time (min)')
plt.ylabel('Distance (km)')

# Setting scalar limits on axes
plt.xlim([0, 40])
plt.ylim([0, 100])

# Plotting values with defined variables
plt.plot(t, d_r, c='green')
plt.plot(t, d_s, c='brown')

# Adding intersection lines
plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple', linestyle='--')
```

### Method 2

##### Editing attributes using plt as well as the ax object returned from plt.subplots()

```python
figure, ax = plt.subplots() # Creates a blank plot with figure and axes as return values that can be modified later

# Creating graph and axes labels
plt.title('Sheriff vs Bank Robber --- Distance vs time graph')
plt.xlabel('Time (min)')
plt.ylabel('Distance (km)')

# Setting scalar limits on axes
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])

# Plotting values with defined variables
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='brown')

# Adding intersection lines
plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```
